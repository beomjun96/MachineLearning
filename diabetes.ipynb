```python
import matplotlib.pylab as plt
from sklearn import linear_model
from sklearn import datasets
reg = linear_model.LinearRegression()

diabetes= datasets.load_diabetes()

num = len(diabetes.data)
num = round(num*0.8)

train_x = diabetes.data[:num]
#test_x = diabetes.target[num:]
test_x = diabetes.data[num:]
# train_y = diabetes.data[:num]
train_y = diabetes.target[:num]
test_y = diabetes.target[num:]

reg.fit(train_x, train_y)
preds = reg.predict(test_x)
error_sum = 0
for idx in range(len(preds)):
  error_sum += (test_y[idx] - preds[idx]) ** 2
print("MSE score:", (error_sum / len(preds)))
```
